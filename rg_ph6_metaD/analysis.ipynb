# Plot Rg

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rg_ph74 = np.loadtxt('rg_ph74.xvg',skiprows=28)
rg_ph6 = np.loadtxt('rg_ph6.xvg',skiprows=28)
```

```python
plt.figure(figsize=(8,6))
plt.rcParams['font.size'] = '16'
sns.kdeplot(rg_ph74[:,1])
sns.kdeplot(rg_ph6[:,1])
plt.xlabel('Radius of Gyration (nm)')
plt.legend(['pH 7.4', 'pH 6'])
plt.title('TRP-ZIP-8')
plt.savefig('rg-ph_distribution.png',dpi=300,bbox_inches='tight',facecolor='white')
```

# Plot intramolecular contacts

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x_ph74 = np.loadtxt('numcount_ph74.xvg',skiprows=28)
x_ph6 = np.loadtxt('numcount_ph6.xvg',skiprows=28)
```

```python
plt.figure(figsize=(8,6))
plt.rcParams['font.size'] = '16'
sns.kdeplot(x_ph74[:,1])
sns.kdeplot(x_ph6[:,1])
plt.xlabel('# Intramolecular contacts', fontsize='16')
plt.legend(['pH 7.4', 'pH 6'])
plt.title('TRP-ZIP-8')
plt.savefig('numcount-ph_distribution.png',dpi=300,bbox_inches='tight',facecolor='white')
```
